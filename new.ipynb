```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os


folder = "E:\\E-Commerce\\E-Commerce Project"
Lf = os.listdir("E:\\E-Commerce\\E-Commerce Project")

```
